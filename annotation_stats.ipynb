```python
import xml.etree.ElementTree as ET
```

```python
# Get the root of the XML file
tree = ET.parse("annotations.xml")
root = tree.getroot()
```

```python
# Explore children in XML file
for child in root:
    print(f"Tag: {child.tag} | Text: {child.attrib}")
```

```python
# Creating variables
project_info = root[1][0]
labels = [label[0].text for label in project_info[9]]
images = root.findall("image")
```

```python
# Create function to count labels

def count_labels(label_name: str, imgs: list) -> tuple[str, int]:
    count = 0

    for img in imgs:
        for label in img:
            if label.get("label") == label_name:
                count += 1

    return (label_name, count)
```

```python
# Output results (Pretty output)
total_annotations = []

for label in labels:
    label_name, count = count_labels(label, images)
    print(f"{label_name}: {count}")
    total_annotations.append(count)

print(
    f"\nTotal annotations: {sum(total_annotations)}"
    f"\n# of images annotated: {len(images)}"
)
```
